```
%%R 
df <- read.csv("glacier.csv", na.strings = "..", comment.char = "#")
str(df)
```

```
%%R 
library("dplyr")
```

```
%%R
df %>% 
  group_by(GEO) %>%
  summarise(Val = max(Ref_Date) - min(Ref_Date)) %>% 
  filter(Val == min(Val)) %>% 
  select(GEO)
```

```
%%R 
df %>%
  filter(MEASURE == "Annual mass balance") %>% 
  group_by(GEO) %>% 
  summarise(M = median(Value, na.rm = T)) %>% 
  filter(abs(M) == min(abs(M))) %>% 
  select(GEO)
```

```
%%R 
df %>% 
  filter(is.na(Value)) %>% 
  select(GEO)
```
