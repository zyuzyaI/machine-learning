```
%%R 
library("ggplot2")
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() 
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() + 
    scale_x_continuous(name = "Miles")
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() + 
    scale_x_continuous(name = "Miles",
                       breaks = c(10, 20, 30, 31, 32, 33))
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() + 
    scale_x_continuous(name = "Miles", 
                       limits = c(1, 35))+
    scale_y_continuous(limits = c(50, 400)) + 
    scale_color_discrete(name = "Legend name")
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() + 
    scale_x_continuous(name = "Miles", 
                       limits = c(1, 35))+
    scale_y_continuous(limits = c(50, 400)) + 
    scale_color_discrete(name = "Legend name",
                         labels = c("Auto", "Manual"))
```

```
%%R 
ggplot(mtcars, aes(x = mpg, y = hp, col = factor(am))) + 
    geom_point() + 
    scale_x_continuous(name = "Miles", 
                       limits = c(1, 35))+
    scale_y_continuous(limits = c(50, 400)) + 
    scale_color_manual(values = c("Yellow", "Blue"),
                        name = "Legend name",
                       labels = c("Auto", "Manual"))
```

```
%%R 
ggplot(mtcars, aes(hp, fill = factor(am))) + 
    geom_density(alpha = 0.2) + 
    scale_fill_discrete(name = "Legend")
```

```
%%R 
ggplot(mtcars, aes(hp, fill = factor(am))) + 
    geom_density(alpha = 0.2) + 
    scale_fill_manual(values = c("Yellow", "Green"),
                      name = "Legend",
                      labels = c("Auto", "Manual"))
```
