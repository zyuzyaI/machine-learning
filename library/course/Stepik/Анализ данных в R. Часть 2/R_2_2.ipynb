````
%%R 
gr_mtcars <- group_by(mtcars, am)
se_data <- summarise(gr_mtcars,
                     mean_mpg = mean(mpg),
                     y_max = mean(mpg) + 1.96 * sd(mpg) / sqrt(length(mpg)),
                     y_min = mean(mpg) - 1.96 * sd(mpg) / sqrt(length(mpg))
                     )
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) +
    geom_errorbar(aes(ymin = y_min, ymax = y_max))
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) + 
    geom_errorbar(aes(ymin = y_min, ymax = y_max), width = 0.2)
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) + 
    geom_errorbar(aes(ymin = y_min, ymax = y_max), width = 0.2) + 
    geom_point(size = 3)
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) + 
    geom_errorbar(aes(ymin = y_min, ymax = y_max), width = 0.2) + 
    geom_point(shape = 21, size = 5, fill = "white")
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) + 
    geom_pointrange(aes(ymin = y_min, ymax = y_max))
````

````
%%R
ggplot(se_data, aes(x = factor(am), y = mean_mpg)) + 
    geom_pointrange(aes(ymin = y_min, ymax = y_max), color = "red")
````

````
%%R 
gr_mtcars <- group_by(mtcars, am, vs)
se_data <- summarise(gr_mtcars,
                     mean_mpg = mean(mpg),
                     y_max = mean(mpg) + 1.96 * sd(mpg) / sqrt(length(mpg)),
                     y_min = mean(mpg) - 1.96 * sd(mpg) / sqrt(length(mpg))
                     )
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg, col = factor(vs))) + 
    geom_errorbar(aes(ymin = y_min, ymax = y_max), width = 0.2) + 
    geom_point(shape = 21, size = 5, fill = "white")
````

````
%%R 
ggplot(se_data, aes(x = factor(am), y = mean_mpg, 
                    col = factor(vs), group = factor(vs)),
       ) + 
    geom_errorbar(aes(ymin = y_min, ymax = y_max), width = 0.2) + 
    geom_point(shape = 21, size = 5, fill = "white") +
    geom_line()
````
