````
%%R 
library("dplyr")
````

````
%%R 
mtcars <- mutate(mtcars, 
                 am = factor(am, label = c("A", "M")),
                 vs = factor(vs, label = c("V", "S")))
````

````
%%R 
ggplot(mtcars, aes(hp)) + 
    geom_dotplot()
````

````
%%R 
ggplot(mtcars, aes(hp)) + 
    geom_dotplot() +
    facet_grid(am ~ vs)
````

````
%%R 
ggplot(mtcars, aes(hp)) +
    geom_dotplot() +
    facet_grid(vs ~ am, margins = T)
````

````
%%R 
ggplot(mtcars, aes(hp, mpg)) + 
    geom_point() +
    facet_grid( am ~ vs, margins = T)
````

````
%%R 
ggplot(mtcars, aes(hp, mpg)) +
    geom_point(aes(col = factor(cyl))) + 
    facet_grid(am ~ vs, margins = T)
````
