```
%%R 
library("ggplot2")
```

```
%%R
ggplot(d, aes(x = date, y = percent, col = system)) +
    geom_point()
```

```
%%R 
ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    theme_classic()
```

```
%%R 
ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_y_continuous(breaks = c(0, 0.04, 0.08, .11, .15)) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    theme_classic()
```

```
%%R 
ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_y_continuous(breaks = c(0, 0.04, 0.08, .11, .15),
                       limits = c(0, .15)) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    theme_classic()
```

```
%%R 
ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_y_continuous(breaks = c(0, 0.04, 0.08, .11, .15),
                       limits = c(0, .15),
                       labels = scales::percent) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    theme_classic()
```

```
%%R 
ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_y_continuous(breaks = c(0, 0.04, 0.08, .11, .15),
                       limits = c(0, .15),
                       labels = scales::percent) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    xlab("") +
    ylab("") +
    theme_classic()
```

```
%%R 
p <- ggplot(d, aes(date, percent, col = system, group = system)) + 
    geom_point(shape = 21, size = 5) +
    geom_point(shape = 21, size = 4.5) +
    geom_point(shape = 21, size = 4) +
    geom_line() + 
    geom_vline(xintercept = 7.5) +
    scale_y_continuous(breaks = c(0, 0.04, 0.08, .11, .15),
                       limits = c(0, .15),
                       labels = scales::percent) +
    scale_color_manual(values = c("orangered1", "red",
                                  "cyan", "yellow1", "springgreen2")) +
    xlab("") +
    ylab("") +
    theme_classic()
```
