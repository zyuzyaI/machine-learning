Вы также можете потренироваться использовать оператор %>%, дополнив следующий код:
```
my_df <- mtcars %>% 
  select() %>% 
  filter() %>% 
  arrange() %>% 
  slice() %>% 
  rename()
  ```

````
%%R
# data(mtcars)
mtcars %>% 
  select(mpg, am, vs, hp) %>% 
  filter(mpg > 14,  hp > 100) %>%
  arrange(desc(mpg)) %>% 
  slice(1:10) %>% 
  rename("Miles per gallon" = mpg,
         "Gross horsepower" = hp)
````
