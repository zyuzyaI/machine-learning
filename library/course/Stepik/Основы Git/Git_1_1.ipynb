Система контроля версий позволяет:
- [ ] Проверять отзывы пользователей на различные версии программы.
- [ ] Публиковать программу в магазинах приложений.
- [x] Фиксировать версию программы на различных этапах разработки, тем самым создавая историю.
- [x] Отслеживать изменения между разными версиями программы.
- [x] Организовать совместную работу нескольких разработчиков

В случае, когда программистов несколько, система контроля версий выполняет функцию

- [ ] Обсуждения деталей реализации.
- [x] Организации совместной работы нескольких программистов. 
- [ ] Распределения задач между программистами.
- [x] Слияния изменений и разрешения конфликтов в файлах исходных кодов.
- [ ] Организации встреч и митапов.

В централизованной системе контроля версий есть

- [x] Только одна копия репозитория
- [ ] Несколько master-копий и несколько slave-копий.
- [ ] Несколько равнозначных копий репозитория.

Минимальной "единицей" в git (исходя из предыдущего видео) является

- [ ] Ветка
- [x] Коммит
- [ ] Репозиторий
