Склонируйте репозиторий по адресу ﻿﻿https://github.com/OSLL/git_course_example_repo в папку REPO, находящуюся в домашней директории текущего пользователя. В результате, содержимое репозитория должно находиться в REPO.

Подробнее: ﻿git clone

```
git clone https://github.com/OSLL/git_course_example_repo.git REPO
```

Репозиторий https://github.com/OSLL/git_course_example_repo  склонирован в папку $HOME/REPO . Настройте git так, чтобы имя git-пользователя было GitMegauser, а e-mail git_super@example.com

Подробнее: git config

```
git config --global user.name "GitMegauser"
git config --global user.email "git_super@example.com"
```
```
git clone https://github.com/OSLL/git_course_example_repo.git HOME/REPO
```

Репозиторий https://github.com/OSLL/git_course_example_repo склонирован в папку $HOME/REPO. Настройте его локальную конфигурацию так, чтобы можно было выполнять команду push на сервер, не вводя логин и пароль github аккаунта.

Подсказка: необходимо сменить origin на ssh-версию. git config, "Which remote URL should I use? - Github"

1. Заходим по ссылке в тексте задачи и копируем ссылку SSH
2. cd REPO
3. git config --list (смотрим как называется 4. параметр в котором указана ссылка на репозиторий)
4. git config remote.origin.url git@github.com:OSLL/git_course_example_repo.git
