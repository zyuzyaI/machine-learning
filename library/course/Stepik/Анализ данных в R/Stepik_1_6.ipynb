При помощи функции ggplot() или boxplot() постройте график boxplot, используя встроенные в R данные airquality. По оси x отложите номер месяца, по оси y — значения переменной Ozone.

На графике boxplot отдельными точками отображаются наблюдения, отклоняющиеся от 1 или 3 квартиля больше чем на полтора межквартильных размаха. Сколько таких наблюдений присутствует в сентябре (месяц №9)?

Обратите внимание, что для корректного отображения графика ggplot ожидает факторную переменную по оси x.

````
%%R 
library(ggplot2)
````

````
%%R 
airquality$Month <- factor(airquality$Month)
ggplot(airquality, aes(x=Month, y=Ozone)) + 
      geom_boxplot()
````

Укажите, при помощи какого варианта кода мы можем построить следующий график по данным iris:
Гистограмма распределения переменной Sepal.Length, в которой цвет заполнения столбцов гистограммы зависит от значения переменной Species.

Возможно несколько вариантов правильного ответа. 

- [ ] ggplot(iris, aes(Sepal.Length)) + geom_histogram(aes(col = Species))
- [x] ggplot(iris, aes(Sepal.Length, fill = Species)) + geom_histogram()
- [ ] ggplot(iris, aes(Sepal.Length, fill = Species)) + geom_histogram()
- [ ] ggplot(iris, aes(Sepal.Length)) + geom_histogram(fill = Species)
- [x] ggplot(iris, aes(Sepal.Length)) + geom_histogram(aes(fill = Species))

````
%%R
ggplot(iris, aes(Sepal.Length)) +
     geom_histogram(aes(col = Species))
````

````
%%R 
ggplot(iris, aes(Sepal.Length, fill = Species)) + geom_histogram()
````

````
%%R 
ggplot(iris, aes(Sepal.Length, col = Species)) + geom_histogram()
````

````
%%R 
ggplot(iris, aes(Sepal.Length)) + geom_histogram(fill = Species)
````

````
%%R 
ggplot(iris, aes(Sepal.Length)) + geom_histogram(aes(fill = Species))
````

Scatterplot (диаграмма рассеивания), где по оси X будет отложена переменная Sepal.Length,  по оси Y переменная  Sepal.Width. За цвет точек будет отвечать переменная  Species, а за размер точек переменная Petal.Length.

Ярослав написал следующую команду

```
ggplot(aes(Sepal.Length, Sepal.Width, col = Species)) +

 geom_point(iris, size = Petal.Length)
```


Однако построить желаемый график не удается! Укажите, какие ошибки совершил Ярослав и попробуйте построить данный график самостоятельно.

Возможно несколько правильных ответов.

- [ ] Выбран неверный geom, необходимо использовать geom_dotplot()
- [x] Первым аргументом функции ggplot() должны быть данные iris.
- [x] Переменная, отвечающая за размер точек должна быть указана внутри функции aes()
- [ ] Невозможно одновременно задавать и цвет и размер точек, только одно из двух. 
- [ ] Чтобы изменять цвет точек необходимо указать fill = Species.

````
%%R 
ggplot(aes(Sepal.Length, Sepal.Width, col = Species)) +
    geom_point(iris, size = Petal.Length)
````

````
%%R 
ggplot(iris, aes(Sepal.Length, Sepal.Width, col = Species)) +
    geom_point()
````
