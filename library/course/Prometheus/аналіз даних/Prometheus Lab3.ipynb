- [ ] ŷ=-0.5x+3

```
%%R
print(lm(y1 ~ x1, data = anscombe))
print(lm(y1 ~ x2, data = anscombe))
print(lm(y3 ~ x3, data = anscombe))
print(lm(y4 ~ x4, data = anscombe))
```
