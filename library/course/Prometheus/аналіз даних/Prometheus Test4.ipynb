5. Тривалість вагітності є нормально розподіленою величиною з середнім значенням 268 днів та середньоквадратичним відхиленням 15 днів. Нехай випадковим чином обрали 36 вагітних жінок. Знайдіть ймовірність, що середнє значення тривалості їхньої вагітності буде меншим, ніж 263 дні.

- [ ] 98%
- [ ] 40%
- [x] 2%

8. В рамках дослідження використання електроенергії було випадковим чином обрано 30 житлових будинків. Було визначено, що середнє споживання електроенергії за 1 квартал становило 375 кіловат-годин(кВт). Знайдіть межу похибки для рівня довіри 99%, якщо середньоквадратичне відхилення вибірки становить 81 кВт.

- [ ] 9.3
- [ ] 37.2
- [ ] 38.1
- [x] 40.8

```
%%R 
n = 30
alpha = 0.99
t <- qt((1-alpha)/2, df=n-1, lower.tail=FALSE)
res <- t * 81/30**0.5
res
```
