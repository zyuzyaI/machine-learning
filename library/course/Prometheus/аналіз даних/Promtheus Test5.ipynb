10. Виробник шин стверджує, що вони будуть служити принаймі 40 000 км. Тест для 40 шин показав, що середній час служби однієї шини становить 39 750 км, середньоквадратичне відхилення 387 км. Чи можемо ми вважати, що середній час придатності шин відрізняється від задекларованих 40 000 для α=0.05?
